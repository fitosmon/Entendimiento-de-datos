```python
import pandas as pd 
```

```python
df = pd.read_csv('Ventas_totales.csv')
df.head()
```

```python
df1 = df.drop('otros', axis=1)
df1
```

```python
valores_nulos = df1.isnull().sum()
valores_nulos
```

```python
df2 = df1.copy()
df2['subtotal_ventas_alimentos_bebidas'] = df2['subtotal_ventas_alimentos_bebidas'].fillna(round(df1['subtotal_ventas_alimentos_bebidas'].mean()))
df2
```

```python
df2['tarjetas_credito'] = df2['tarjetas_credito'].fillna(round(df1['tarjetas_credito'].mean()))
df2
```

```python
df2['salon_ventas'] = df2['salon_ventas'].fillna(round(df1['salon_ventas'].median()))
df2
```

```python
df2['tarjetas_debito'] = df2['tarjetas_debito'].ffill()
df2
```

```python
df2['electronicos_articulos_hogar'] = df2['electronicos_articulos_hogar'].bfill()
df2
```

```python
df2['otros_medios'] = df2['otros_medios'].ffill()
df2
```

```python
df2['otros_medios'] = df2['otros_medios'].fillna(round(df1['otros_medios'].mean()))
df2
```

```python
df2['almacen'] = df2['almacen'].ffill()
df2
```

```python
df2['bebidas'] = df2['bebidas'].ffill()
df2
```

```python
df2['panaderia'] = df2['panaderia'].bfill()
df2
```

```python
df2['lacteos'] = df2['lacteos'].bfill()
df2
```

```python
df2['carnes'] = df2['carnes'].fillna(round(df1['carnes'].mean()))
df2
```

```python
df2['verduleria_fruteria'] = df2['verduleria_fruteria'].fillna(round(df1['verduleria_fruteria'].median()))
df2
```

```python
df2['alimentos_preparados_rotiseria'] = df2['alimentos_preparados_rotiseria'].fillna(round(df1['alimentos_preparados_rotiseria'].median()))
df2
```

```python
df2['indumentaria_calzado_textiles_hogar'] = df2['indumentaria_calzado_textiles_hogar'].ffill()
df2
```

```python
df2 = df2.bfill()
df2
```

```python
valores_nulos = df2.isnull().sum()
valores_nulos
```

```python
file= open('3-1-valoresNulos.csv', 'a')
file.write("Se sustituyeron los valores nulos de 14 columnas con los metodos de bfill, ffill, mediana y media. Ademas se elimino la ultima columna por parecer innecesario\n")
file.write("No se considero conveniente eliminar filas con el metodo dropna para evitar perdida de informacion \n")
df2.to_csv(file)
file.close()
```
