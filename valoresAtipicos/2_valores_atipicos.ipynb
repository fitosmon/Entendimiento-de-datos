```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("./../Ventas_totales.csv")
data.info()
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
cuantitativas = data.iloc[:, 1:23]
cualitativas = data.iloc[:, [0]]
cualitativas
```

```python
cuantitativas
```

```python
fig = plt.figure(figsize=(15, 8))
#cuantitativas['ventas_precios_corrientes'].plot(kind='box', vert=False)
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores atipicos del DataFrame")
plt.show()
```

PROCEDIMIENTO "DESVIACION ESTANDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME

```python
#metodo aplicando desviacion estandar. Encuentro los valores extremos
y=cuantitativas
Limite_Superior = y.mean() + 3*y.std()
Limite_inferior = y.mean() - 3*y.std()
print("Limite_Superior permitido\n", Limite_Superior)
print("Limite inferior permitido\n", Limite_inferior)
```

```python
data3 = cuantitativas[(y<=Limite_Superior)&(y>=Limite_inferior)]
data3
```

```python
valores_nulos = data3.isnull().sum()
valores_nulos
```

```python
#reemplazamos valores atipicos (nulos) del dataframe con mean
#realizamos una copia del datafram
data_clean = data3.copy()
data_clean = data_clean.fillna(round(data3.mean(), 1))
data_clean
```

```python
valores_nulos = data_clean.isnull().sum()
valores_nulos
```

Procedimiento "cuantiles" para sustituir valores atipicos

```python
y = cuantitativas
percentile25 = y.quantile(0.25)
percentile75 = y.quantile(0.75)
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5 *iqr
Limite_inferior_iqr = percentile25 + 1.5*iqr
print("Limite superior permitido \n", Limite_Superior_iqr)
print("Limite inferior permitido \n", Limite_inferior_iqr)
```

```python
data3_iqr = cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_inferior_iqr)]
data3_iqr
```

```python
valores_nulos = data3_iqr.isnull().sum()
valores_nulos
```

```python
data4_iqr = data3_iqr.copy()
data4_iqr = data4_iqr.fillna(round(data3_iqr.mean(), 1))
data4_iqr
```

```python
Datos_limpios = pd.concat([cualitativas, data4_iqr], axis=1)
Datos_limpios
```

```python
Datos_limpios.to_csv("datosLimpiosClase.csv")
```
